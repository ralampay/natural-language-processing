```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
words = open('names.txt', 'r').read().splitlines()

temp_words = []

for word in words:
    temp_words.append(word.lower())
    
words = temp_words
```

```python
words[:10]
```

```python
len(words)
```

```python
b = {}
for word in words:
    chs = ['<S>'] + list(word.lower()) + ['<E>']
    for ch1, ch2 in zip(chs, chs[1:]):
        bigram = (ch1, ch2)
        b[bigram] = b.get(bigram, 0) + 1
        #print(ch1, ch2)
```

```python
sorted(b.items(), key = lambda kv: -kv[1])
```

```python
N = np.zeros((28, 28)).astype(np.int32)
```

```python
N
```

```python
chars = sorted(list(set(''.join(words))))

len(chars)
```

```python
stoi = {s:i for i,s in enumerate(chars)}
stoi['<S>'] = 26
stoi['<E>'] = 27

stoi
```

```python
for word in words:
    chs = ['<S>'] + list(word) + ['<E>']
    for ch1, ch2 in zip(chs, chs[1:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        
        N[ix1, ix2] += 1
```

```python
N
```

```python
plt.imshow(N)
```

```python
itos = []

for elem in stoi:
    itos.append(elem)
    
itos
```

```python
plt.figure(figsize=(16,16))
plt.imshow(N, cmap='Blues')
for i in range(len(stoi)):
    for j in range(len(stoi)):
        chstr = itos[i] + itos[j]
        plt.text(j, i, chstr, ha="center", va="bottom", color='gray')
        plt.text(j, i, N[i, j], ha="center", va="top", color="gray")
        
plt.axis('off')
```
